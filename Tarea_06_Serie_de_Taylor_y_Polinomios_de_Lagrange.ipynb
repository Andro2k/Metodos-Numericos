### **Ejercicio 1:**
$(\frac{1}{2\cos^2{x}+1})$ en $(x_0 = 0)$

Primero, encontramos la serie de Taylor para $(f(x) = \frac{1}{2\cos^2{x}+1})$ alrededor de $(x = 0)$.

1. $(f(x) = \frac{1}{2\cos^2{x}+1})$
2. $(f(0) = \frac{1}{2\cos^2{0}+1} = \frac{1}{2 $cdot 1 + 1} = \frac{1}{3})$
3. Calculamos las derivadas de $(f(x))$ y evaluamos en $(x = 0)$.

#### Derivadas de $(f(x) = \frac{1}{2\cos^2{x}+1})$

La función $(f(x))$ y sus derivadas se pueden encontrar usando las reglas de derivación. Dado que puede ser un cálculo extenso, utilizamos una herramienta computacional para obtener estas derivadas y los polinomios resultantes.

### Serie de Taylor de $(f(x))$

La serie de Taylor de la función $( f(x) = \frac{1}{2\cos^2{x} + 1} )$ alrededor de $( x = 0 )$ hasta el término de $( x^4 )$ es:

$
[ f(x) \approx \frac{1}{3} + \frac{2}{9}x^2 + \frac{2}{27}x^4
$

---

```python
from sympy import symbols, cos, series

# Define the variable and the function
x = symbols('x')
f = 1 / (2*cos(x)**2 + 1)

# Calculate the Taylor series around x0 = 0
taylor_series_f = series(f, x, 0, 5)  # Up to x^4 for better approximation

taylor_series_f
```

### **Graficar las diferentes aproximaciones**

Ahora, graficamos las funciones originales y sus aproximaciones utilizando los polinomios de Taylor encontrados.

#### Gráficas de las Aproximaciones

1. **Función $( \frac{1}{2\cos^2{x} + 1})$** en $( x = 0 )$:
   - **Función Original**: $( \frac{1}{2\cos^2{x} + 1} )$
   - **Serie de Taylor**: $( \frac{1}{3} + \frac{2}{9}x^2 + \frac{2}{27}x^4 )$
   - La gráfica muestra que la serie de Taylor se ajusta bien a la función original cerca de $( x = 0 )$.

2. **Función $( \arctan{x} )$** en $( x = 1 )$:
   - **Función Original**: $( \arctan{x} )$
   - **Serie de Taylor**: $( \frac{\pi}{4} + \frac{x-1}{2} - \frac{(x-1)^2}{4} + \frac{(x-1)^3}{12} )$
   - La gráfica muestra que la serie de Taylor se ajusta bien a la función original cerca de $( x = 1 )$.

Estas aproximaciones permiten entender cómo se comportan las funciones alrededor de los puntos de expansión dados. La serie de Taylor proporciona una forma efectiva de aproximar funciones complicadas mediante polinomios más simples.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original functions and their Taylor approximations
def original_f1(x):
    return 1 / (2*np.cos(x)**2 + 1)

def taylor_f1(x):
    return 1/3 + (2/9)*x**2 + (2/27)*x**4

def original_f2(x):
    return np.arctan(x)

def taylor_f2(x):
    return (np.pi/4) + (x-1)/2 - (x-1)**2/4 + (x-1)**3/12

# Define the range for plotting
x1 = np.linspace(-1, 1, 400)
x2 = np.linspace(0, 2, 400)

# Plotting
plt.figure(figsize=(12, 6))

# Plot for the first function
plt.subplot(1, 2, 1)
plt.plot(x1, original_f1(x1), label='Original $\\frac{1}{2\\cos^2{x} + 1}$')
plt.plot(x1, taylor_f1(x1), label='Series de Taylor$\\frac{1}{3} + \\frac{2}{9}x^2 + \\frac{2}{27}x^4$', linestyle='--')
plt.title('Aproximacion de $\\frac{1}{2\\cos^2{x} + 1}$ cuando $x=0$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)

# Plot for the second function
plt.subplot(1, 2, 2)
plt.plot(x2, original_f2(x2), label='Original $\\arctan{x}$')
plt.plot(x2, taylor_f2(x2), label='Series de Taylor $\\frac{\\pi}{3} + \\frac{x-1}{2} - \\frac{(x-1)^2}{4} + \\frac{(x-1)^3}{12}$', linestyle='--')
plt.title('Aproximacion de $\\arctan{x}$ cuando $x=1$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
