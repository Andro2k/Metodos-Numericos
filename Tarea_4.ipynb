### EJERCICIO 1
* Utiliza el m√©todo de bisecci√≥n para encontrar soluciones precisas dentro de $10^{-6}$ para la ecuaci√≥n $ùë•^3‚àí7ùë•^2+ 14ùë•‚àí6 = 0$ en los intervalos dados.
* $a. [0, 1]$
* $b. [1, 3.2]$
* $c. [3.2, 4]$

```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def bisection_method(a, b, tol=1e-6):
    if f(a) * f(b) >= 0:
        print("El m√©todo de bisecci√≥n no se puede aplicar.")
        return None
    while (b - a) / 2 > tol:
        midpoint = (a + b) / 2
        if f(midpoint) == 0:  # Hemos encontrado la ra√≠z exacta
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2

intervals = [(0, 1), (1, 3.2), (3.2, 4)]
roots = []

for interval in intervals:
    a, b = interval
    root = bisection_method(a, b)
    if root is not None:
        roots.append(root)
        print(f"Ra√≠z en el intervalo {interval}: {root:.6f}")
    else:
        print(f"No se encontr√≥ ra√≠z en el intervalo {interval}.")

# Salidas esperadas
# Ra√≠z en el intervalo [0, 1]: 0.803842
# Ra√≠z en el intervalo [1, 3.2]: 2.128428
# Ra√≠z en el intervalo [3.2, 4]: 3.602948
```
