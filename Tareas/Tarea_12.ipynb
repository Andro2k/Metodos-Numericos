## **Ejercicio 1:**

Use el método de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.

a. $( y' = t e^{3t} - 2y )$, $( 0 \leq t \leq 1 )$, $( y(0) = 0 )$, con $( h = 0.5 )$

b. $( y' = 1 + (t - y)^2 )$, $( 2 \leq t \leq 3 )$, $( y(2) = 1 )$, con $( h = 0.5 )$

c. $( y' = 1 + \frac{y}{t} )$, $( 1 \leq t \leq 2 )$, $( y(1) = 2 )$, con $( h = 0.25 )$

d. $( y' = \cos(2t) + \sin(3t) )$, $( 0 \leq t \leq 1 )$, $( y(0) = 1 )$, con $( h = 0.25 )$

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, t0, y0, t_final, h):
    """
    Método de Euler para aproximar la solución de una ODE de primer orden.

    f: función que representa la derivada y'(t, y)
    t0: valor inicial de t
    y0: valor inicial de y
    t_final: valor final de t
    h: tamaño del paso

    Retorna una lista de valores t y una lista de valores y aproximados.
    """
    n = int((t_final - t0) / h) + 1
    t_values = np.linspace(t0, t_final, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])

    return t_values, y_values

def plot_results(t_values, y_values, title):
    """
    Función para graficar los resultados del método de Euler.

    t_values: lista de valores t
    y_values: lista de valores y
    title: título del gráfico
    """
    plt.plot(t_values, y_values, 'o-', label='Aproximación (Método de Euler)')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Definimos la función derivada para el primer caso
def f_a(t, y):
    return t * np.exp(3 * t) - 2 * y

# Resolviendo con el método de Euler
t_values_a, y_values_a = euler_method(f_a, t0=0, y0=0, t_final=1, h=0.5)

# Graficamos los resultados
plot_results(t_values_a, y_values_a, 'a) Solución para y\' = t $e^{3t}$ - 2y')
```

```python
# Definimos la función derivada para el segundo caso
def f_b(t, y):
    return 1 + (t - y)**2

# Resolviendo con el método de Euler
t_values_b, y_values_b = euler_method(f_b, t0=2, y0=1, t_final=3, h=0.5)

# Graficamos los resultados
plot_results(t_values_b, y_values_b, 'b) Solución para y\'$ = 1 + (t - y)^2$')
```

```python
# Definimos la función derivada para el tercer caso
def f_c(t, y):
    return 1 + y / t

# Resolviendo con el método de Euler
t_values_c, y_values_c = euler_method(f_c, t0=1, y0=2, t_final=2, h=0.25)

# Graficamos los resultados
plot_results(t_values_c, y_values_c, 'c) Solución para y\' = 1 + y/t')
```

```python
# Definimos la función derivada para el cuarto caso
def f_d(t, y):
    return np.cos(2 * t) + np.sin(3 * t)

# Resolviendo con el método de Euler
t_values_d, y_values_d = euler_method(f_d, t0=0, y0=1, t_final=1, h=0.25)

# Graficamos los resultados
plot_results(t_values_d, y_values_d, 'd) Solución para y\' = cos(2t) + sin(3t)')
```

b. $( y(t) = t + \frac{1}{1-t} )$

c. $( y(t) = t \ln(t) + 2t )$

d. $( y(t) = \frac{1}{2}\sin(2t) - \frac{1}{3}\cos(3t) + \frac{4}{3} )$

```python
# Soluciones exactas dadas

def exact_a(t):
    return (1/5) * t * np.exp(3 * t) - (1/25) * np.exp(3 * t) + (1/25) * np.exp(-2 * t)

def exact_b(t):
    return t + 1/(1 - t)

def exact_c(t):
    return t * np.log(t) + 2 * t

def exact_d(t):
    return 0.5 * np.sin(2 * t) - (1/3) * np.cos(3 * t) + 4/3
```

```python
def compute_error(exact_solution, t_values, y_values):
    """
    Calcula el error real en cada paso.

    exact_solution: función de la solución exacta
    t_values: lista de valores t donde se calculó la aproximación
    y_values: lista de valores y aproximados usando el método de Euler

    Retorna una lista con los errores en cada paso.
    """
    exact_values = exact_solution(t_values)
    error_values = np.abs(exact_values - y_values)
    return error_values

def print_error(t_values, y_values, exact_values, error_values):
    """
    Imprime el valor aproximado, el valor exacto y el error en cada paso.
    """
    for i in range(len(t_values)):
        print(f"t = {t_values[i]:.2f}, y_approx = {y_values[i]:.4f}, y_exact = {exact_values[i]:.4f}, error = {error_values[i]:.4f}")

# Resolver el problema (a) y calcular el error
t_values_a, y_values_a = euler_method(f_a, t0=0, y0=0, t_final=1, h=0.5)
error_values_a = compute_error(exact_a, t_values_a, y_values_a)
print("Errores para el problema (a):")
print_error(t_values_a, y_values_a, exact_a(t_values_a), error_values_a)
plot_results(t_values_a, error_values_a, 'Error para y\' = t e^{3t} - 2y')

# Resolver el problema (b) y calcular el error
t_values_b, y_values_b = euler_method(f_b, t0=2, y0=1, t_final=3, h=0.5)
error_values_b = compute_error(exact_b, t_values_b, y_values_b)
print("\nErrores para el problema (b):")
print_error(t_values_b, y_values_b, exact_b(t_values_b), error_values_b)
plot_results(t_values_b, error_values_b, 'Error para y\' = 1 + (t - y)^2')

# Resolver el problema (c) y calcular el error
t_values_c, y_values_c = euler_method(f_c, t0=1, y0=2, t_final=2, h=0.25)
error_values_c = compute_error(exact_c, t_values_c, y_values_c)
print("\nErrores para el problema (c):")
print_error(t_values_c, y_values_c, exact_c(t_values_c), error_values_c)
plot_results(t_values_c, error_values_c, 'Error para y\' = 1 + y/t')

# Resolver el problema (d) y calcular el error
t_values_d, y_values_d = euler_method(f_d, t0=0, y0=1, t_final=1, h=0.25)
error_values_d = compute_error(exact_d, t_values_d, y_values_d)
print("\nErrores para el problema (d):")
print_error(t_values_d, y_values_d, exact_d(t_values_d), error_values_d)
plot_results(t_values_d, error_values_d, 'Error para y\' = cos(2t) + sin(3t)')
```

## **Ejercicio 3:**

Utilice el método de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.

a. $( y' = \frac{y}{t} - \left(\frac{y}{t}\right)^2 )$, $( 1 \leq t \leq 2 )$, $( y(1) = 1 )$, con $( h = 0.1 )$

b. $( y' = 1 + \frac{y}{t} + \left(\frac{y}{t}\right)^2 )$, $( 1 \leq t \leq 3 )$, $( y(1) = 0 )$, con $( h = 0.2 )$

c. $( y' = -(y + 1)(y + 3) )$, $( 0 \leq t \leq 2 )$, $( y(0) = -2 )$, con $( h = 0.2 )$

d. $( y' = -5y + 5t^2 + 2t )$, $( 0 \leq t \leq 1 )$, $( y(0) = \frac{1}{3} )$, con $( h = 0.1 )$

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, t0, y0, t_final, h):
    """
    Aplica el método de Euler para aproximar la solución de una ODE.

    f: función que define la derivada dy/dt
    t0: valor inicial de t
    y0: valor inicial de y
    t_final: valor final de t hasta donde se quiere aproximar
    h: tamaño del paso

    Retorna una lista de valores t y una lista de valores y aproximados.
    """
    n = int((t_final - t0) / h) + 1
    t_values = np.linspace(t0, t_final, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])

    return t_values, y_values

def plot_results(t_values, y_values, title):
    """
    Grafica los resultados de la aproximación.
    """
    plt.plot(t_values, y_values, 'o-', label='Aproximación (Método de Euler)')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Problema (a)
def f_a(t, y):
    return (y / t) - (y / t)**2

t_values_a, y_values_a = euler_method(f_a, t0=1, y0=1, t_final=2, h=0.1)
plot_results(t_values_a, y_values_a, "Aproximación para y' = y/t - $(y/t)^2$")

# Problema (b)
def f_b(t, y):
    return 1 + (y / t) + (y / t)**2

t_values_b, y_values_b = euler_method(f_b, t0=1, y0=0, t_final=3, h=0.2)
plot_results(t_values_b, y_values_b, "Aproximación para y' = 1 + y/t + $(y/t)^2$")

# Problema (c)
def f_c(t, y):
    return -(y + 1) * (y + 3)

t_values_c, y_values_c = euler_method(f_c, t0=0, y0=-2, t_final=2, h=0.2)
plot_results(t_values_c, y_values_c, "Aproximación para y' = -$(y + 1)(y + 3)$")

# Problema (d)
def f_d(t, y):
    return -5 * y + 5 * t**2 + 2 * t

t_values_d, y_values_d = euler_method(f_d, t0=0, y0=1/3, t_final=1, h=0.1)
plot_results(t_values_d, y_values_d, "Aproximación para y' = $-5y + 5t^2 + 2t$")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Euler
def euler_method(f, t0, y0, t_final, h):
    n = int((t_final - t0) / h) + 1
    t_values = np.linspace(t0, t_final, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])

    return t_values, y_values

# Función para calcular el error
def compute_error(exact_solution, t_values, y_values):
    exact_values = exact_solution(t_values)
    return np.abs(exact_values - y_values)

# Función para imprimir los errores
def print_error(t_values, y_values, exact_values, error_values):
    print(f"{'t':>5} {'y_aprox':>10} {'y_exact':>10} {'error':>10}")
    print("-" * 35)
    for t, y_aprox, y_exact, error in zip(t_values, y_values, exact_values, error_values):
        print(f"{t:>5.2f} {y_aprox:>10.6f} {y_exact:>10.6f} {error:>10.6f}")

# Soluciones exactas
def exact_a(t):
    return t / (1 + np.log(t))

def exact_b(t):
    return t * np.tan(np.log(t))

def exact_c(t):
    return -3 + 2 / (1 + np.exp(-2 * t))

def exact_d(t):
    return t**2 + (1/3) * np.exp(-5 * t)

# Problemas (a), (b), (c) y (d)
problems = [
    {
        "f": lambda t, y: (y / t) - (y / t)**2,
        "t0": 1, "y0": 1, "t_final": 2, "h": 0.1,
        "exact_solution": exact_a, "title": "Problema (a)"
    },
    {
        "f": lambda t, y: 1 + (y / t) + (y / t)**2,
        "t0": 1, "y0": 0, "t_final": 3, "h": 0.2,
        "exact_solution": exact_b, "title": "Problema (b)"
    },
    {
        "f": lambda t, y: -(y + 1) * (y + 3),
        "t0": 0, "y0": -2, "t_final": 2, "h": 0.2,
        "exact_solution": exact_c, "title": "Problema (c)"
    },
    {
        "f": lambda t, y: -5 * y + 5 * t**2 + 2 * t,
        "t0": 0, "y0": 1/3, "t_final": 1, "h": 0.1,
        "exact_solution": exact_d, "title": "Problema (d)"
    }
]

# Resolución de los problemas y cálculo de errores
for problem in problems:
    t_values, y_values = euler_method(
        problem["f"], problem["t0"], problem["y0"], problem["t_final"], problem["h"]
    )
    error_values = compute_error(problem["exact_solution"], t_values, y_values)
    print(f"Errores para {problem['title']}:")
    print_error(t_values, y_values, problem["exact_solution"](t_values), error_values)
    print("\n")
```

## **Ejercicio 5:**

Utilice los resultados del ejercicio 3 y la interpolación lineal para aproximar los siguientes valores de $( y(t) )$. Compare las aproximaciones asignadas para los valores reales obtenidos mediante las funciones determinadas en el ejercicio 4.

a. $( y(0.25) )$ y $( y(0.93) )$

b. $( y(1.25) )$ y $( y(1.93) )$

c. $( y(2.10) )$ y $( y(2.75) )$

d. $( y(0.54) )$ y $( y(0.94) )$

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Euler (Ejercicio 4)
def euler_method(f, t0, y0, t_final, h):
    n = int((t_final - t0) / h) + 1
    t_values = np.linspace(t0, t_final, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])

    return t_values, y_values

# Función para la interpolación lineal (Ejercicio 3)
def linear_interpolation(t, t_values, y_values):
    for i in range(len(t_values) - 1):
        if t_values[i] <= t <= t_values[i + 1]:
            return y_values[i] + (y_values[i + 1] - y_values[i]) * (t - t_values[i]) / (t_values[i + 1] - t_values[i])
    return None

# Función para calcular el error
def compute_error(exact_solution, t_values, y_values):
    exact_values = exact_solution(t_values)
    return np.abs(exact_values - y_values)

# Soluciones exactas (Ejercicio 4)
def exact_a(t):
    return t / (1 + np.log(t))

def exact_b(t):
    return t * np.tan(np.log(t))

def exact_c(t):
    return -3 + 2 / (1 + np.exp(-2 * t))

def exact_d(t):
    return t**2 + (1/3) * np.exp(-5 * t)

# Definir los problemas y resolverlos (Ejercicio 4)
problems = [
    {
        "f": lambda t, y: (y / t) - (y / t)**2,
        "t0": 1, "y0": 1, "t_final": 2, "h": 0.1,
        "exact_solution": exact_a, "title": "Problema (a)"
    },
    {
        "f": lambda t, y: 1 + (y / t) + (y / t)**2,
        "t0": 1, "y0": 0, "t_final": 3, "h": 0.2,
        "exact_solution": exact_b, "title": "Problema (b)"
    },
    {
        "f": lambda t, y: -(y + 1) * (y + 3),
        "t0": 0, "y0": -2, "t_final": 2, "h": 0.2,
        "exact_solution": exact_c, "title": "Problema (c)"
    },
    {
        "f": lambda t, y: -5 * y + 5 * t**2 + 2 * t,
        "t0": 0, "y0": 1/3, "t_final": 1, "h": 0.1,
        "exact_solution": exact_d, "title": "Problema (d)"
    }
]

# Resolver cada problema y hacer la interpolación (Ejercicio 3)
for problem in problems:
    t_values, y_values = euler_method(
        problem["f"], problem["t0"], problem["y0"], problem["t_final"], problem["h"]
    )

    # Puntos específicos donde queremos interpolar y comparar
    t_points = [0.25, 0.93, 1.25, 1.93, 2.10, 2.75, 0.54, 0.94]

    # Filtrar puntos según el rango del problema actual
    t_points_filtered = [t for t in t_points if t_values[0] <= t <= t_values[-1]]

    # Realizar interpolación lineal
    approx_values = [linear_interpolation(t, t_values, y_values) for t in t_points_filtered]

    # Valores exactos para comparación
    real_values = [problem["exact_solution"](t) for t in t_points_filtered]

    # Mostrar resultados
    print(f"Comparación de aproximaciones e interpolaciones para {problem['title']}:")
    for t, approx, real in zip(t_points_filtered, approx_values, real_values):
        print(f"y({t:.2f}) - Aproximado: {approx:.4f}, Real: {real:.4f}, Error: {abs(approx - real):.4f}")
    print("\n")

# Opcional: Visualización de resultados
plt.plot(t_values, y_values, 'o-', label='Euler')
plt.plot(t_points_filtered, approx_values, 'x', label='Interpolaciones')
plt.legend()
plt.show()
```

## **Ejercicio 6:**

Use el método de Taylor de orden 2 para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial.

a. $( y' = t e^{3t} - 2y )$, $( 0 \leq t \leq 1 )$, $( y(0) = 0 )$, con $( h = 0.5 )$

b. $( y' = 1 + (t - y)^2 )$, $( 2 \leq t \leq 3 )$, $( y(2) = 1 )$, con $( h = 0.5 )$

c. $( y' = 1 + \frac{y}{t} )$, $( 1 \leq t \leq 2 )$, $( y(1) = 2 )$, con $( h = 0.25 )$

d. $( y' = \cos(2t) + \sin(3t) )$, $( 0 \leq t \leq 1 )$, $( y(0) = 1 )$, con $( h = 0.25 )$


```python
import numpy as np

# Método de Taylor de orden 2
def taylor_order_2(f, f_prime, t0, y0, t_final, h):
    n = int((t_final - t0) / h) + 1
    t_values = np.linspace(t0, t_final, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1]) + \
                      (h**2 / 2) * f_prime(t_values[i-1], y_values[i-1])

    return t_values, y_values

# Problema (a)
f_a = lambda t, y: t * np.exp(3 * t) - 2 * y
f_prime_a = lambda t, y: 3 * t * np.exp(3 * t) - 2 * f_a(t, y)

t_values_a, y_values_a = taylor_order_2(f_a, f_prime_a, 0, 0, 1, 0.5)
print(f"Problema (a): t = {t_values_a}, y = {y_values_a}")

# Problema (b)
f_b = lambda t, y: 1 + (t - y)**2
f_prime_b = lambda t, y: 2 * (t - y) * (1 - f_b(t, y))

t_values_b, y_values_b = taylor_order_2(f_b, f_prime_b, 2, 1, 3, 0.5)
print(f"Problema (b): t = {t_values_b}, y = {y_values_b}")

# Problema (c)
f_c = lambda t, y: 1 + y / t
f_prime_c = lambda t, y: (1 / t) * (1 - f_c(t, y))

t_values_c, y_values_c = taylor_order_2(f_c, f_prime_c, 1, 2, 2, 0.25)
print(f"Problema (c): t = {t_values_c}, y = {y_values_c}")

# Problema (d)
f_d = lambda t, y: np.cos(2 * t) + np.sin(3 * t)
f_prime_d = lambda t, y: -2 * np.sin(2 * t) + 3 * np.cos(3 * t)

t_values_d, y_values_d = taylor_order_2(f_d, f_prime_d, 0, 1, 1, 0.25)
print(f"Problema (d): t = {t_values_d}, y = {y_values_d}")
```

Repita el ejercicio 6 con el método de Taylor de orden 4.

```python
import numpy as np

# Método de Taylor de orden 4
def taylor_order_4(f, f_prime, f_double_prime, f_triple_prime, t0, y0, t_final, h):
    n = int((t_final - t0) / h) + 1
    t_values = np.linspace(t0, t_final, n)
    y_values = np.zeros(n)
    y_values[0] = y0

    for i in range(1, n):
        y_values[i] = (
            y_values[i-1]
            + h * f(t_values[i-1], y_values[i-1])
            + (h**2 / 2) * f_prime(t_values[i-1], y_values[i-1])
            + (h**3 / 6) * f_double_prime(t_values[i-1], y_values[i-1])
            + (h**4 / 24) * f_triple_prime(t_values[i-1], y_values[i-1])
        )

    return t_values, y_values

# Problema (a)
f_a = lambda t, y: t * np.exp(3 * t) - 2 * y
f_prime_a = lambda t, y: 3 * t * np.exp(3 * t) - 2 * f_a(t, y)
f_double_prime_a = lambda t, y: 9 * t * np.exp(3 * t) - 2 * f_prime_a(t, y)
f_triple_prime_a = lambda t, y: 27 * t * np.exp(3 * t) - 2 * f_double_prime_a(t, y)

t_values_a, y_values_a = taylor_order_4(f_a, f_prime_a, f_double_prime_a, f_triple_prime_a, 0, 0, 1, 0.5)
print(f"Problema (a): t = {t_values_a}, y = {y_values_a}")

# Problema (b)
f_b = lambda t, y: 1 + (t - y)**2
f_prime_b = lambda t, y: 2 * (t - y) * (1 - f_b(t, y))
f_double_prime_b = lambda t, y: 2 * (1 - f_b(t, y))**2 - 2 * f_prime_b(t, y)
f_triple_prime_b = lambda t, y: -4 * (1 - f_b(t, y)) * f_prime_b(t, y) - 2 * f_double_prime_b(t, y)

t_values_b, y_values_b = taylor_order_4(f_b, f_prime_b, f_double_prime_b, f_triple_prime_b, 2, 1, 3, 0.5)
print(f"Problema (b): t = {t_values_b}, y = {y_values_b}")

# Problema (c)
f_c = lambda t, y: 1 + y / t
f_prime_c = lambda t, y: (1 / t) * (1 - f_c(t, y))
f_double_prime_c = lambda t, y: (-1 / t**2) * (1 - f_c(t, y)) - (1 / t) * f_prime_c(t, y)
f_triple_prime_c = lambda t, y: (2 / t**3) * (1 - f_c(t, y)) + (2 / t**2) * f_prime_c(t, y) - (1 / t) * f_double_prime_c(t, y)

t_values_c, y_values_c = taylor_order_4(f_c, f_prime_c, f_double_prime_c, f_triple_prime_c, 1, 2, 2, 0.25)
print(f"Problema (c): t = {t_values_c}, y = {y_values_c}")

# Problema (d)
f_d = lambda t, y: np.cos(2 * t) + np.sin(3 * t)
f_prime_d = lambda t, y: -2 * np.sin(2 * t) + 3 * np.cos(3 * t)
f_double_prime_d = lambda t, y: -4 * np.cos(2 * t) - 9 * np.sin(3 * t)
f_triple_prime_d = lambda t, y: 8 * np.sin(2 * t) - 27 * np.cos(3 * t)

t_values_d, y_values_d = taylor_order_4(f_d, f_prime_d, f_double_prime_d, f_triple_prime_d, 0, 1, 1, 0.25)
print(f"Problema (d): t = {t_values_d}, y = {y_values_d}")
```
