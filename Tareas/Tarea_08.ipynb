## **1. Dados los datos:**

| xi  | 4.0  | 4.2  | 4.5  | 4.7  | 5.1  | 5.5  | 5.9  | 6.3  | 6.8  | 7.1  |
| --- | ---- | ---- | ---- | ---- | ---- | ---- | ---- | ---- | ---- | ---- |
| yi  | 102.56 | 130.11 | 113.18 | 142.05 | 167.53 | 195.14 | 224.87 | 256.73 | 299.50 | 326.72 |

- a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error.  
- b. Construya el polinomio por mínimos cuadrados de grado 2 y calcule el error.  
- c. Construya el polinomio por mínimos cuadrados de grado 3 y calcule el error.  
- d. Construya el polinomio por mínimos cuadrados de la forma $(be^{ax})$ y calcule el error.  
- e. Construya el polinomio por mínimos cuadrados de la forma $(bx^a)$ y calcule el error.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def ajustar_polinomios(xi, yi):
    # Ajuste de polinomio de grado 1
    p1 = np.polyfit(xi, yi, 1)
    y1_fit = np.polyval(p1, xi)
    error1 = np.sum((yi - y1_fit)**2)

    # Ajuste de polinomio de grado 2
    p2 = np.polyfit(xi, yi, 2)
    y2_fit = np.polyval(p2, xi)
    error2 = np.sum((yi - y2_fit)**2)

    # Ajuste de polinomio de grado 3
    p3 = np.polyfit(xi, yi, 3)
    y3_fit = np.polyval(p3, xi)
    error3 = np.sum((yi - y3_fit)**2)

    # Función para el ajuste de la forma be^x
    def func_exp(x, b, a):
        return b * np.exp(a * x)

    popt_exp, _ = curve_fit(func_exp, xi, yi, p0=(1, 0.1))
    y_exp_fit = func_exp(xi, *popt_exp)
    error_exp = np.sum((yi - y_exp_fit)**2)

    # Función para el ajuste de la forma b * x^a
    def func_power(x, b, a):
        return b * x**a

    popt_power, _ = curve_fit(func_power, xi, yi, p0=(1, 1))
    y_power_fit = func_power(xi, *popt_power)
    error_power = np.sum((yi - y_power_fit)**2)

    # Mostrar resultados
    print(f"Polinomio de grado 1: Coeficientes: {p1}, Error: {error1}")
    print(f"Polinomio de grado 2: Coeficientes: {p2}, Error: {error2}")
    print(f"Polinomio de grado 3: Coeficientes: {p3}, Error: {error3}")
    print(f"Ajuste exponencial: Coeficientes: {popt_exp}, Error: {error_exp}")
    print(f"Ajuste de potencia: Coeficientes: {popt_power}, Error: {error_power}")

    # Graficar resultados
    plt.scatter(xi, yi, label='Datos')
    plt.plot(xi, y1_fit, label='Polinomio Grado 1', linestyle='--')
    plt.plot(xi, y2_fit, label='Polinomio Grado 2', linestyle='--')
    plt.plot(xi, y3_fit, label='Polinomio Grado 3', linestyle='--')
    plt.plot(xi, y_exp_fit, label='Ajuste Exponencial', linestyle='--')
    plt.plot(xi, y_power_fit, label='Ajuste de Potencia', linestyle='--')
    plt.legend()
    plt.xlabel('xi')
    plt.ylabel('yi')
    plt.title('Ajustes por Mínimos Cuadrados')
    plt.grid(True)
    plt.show()
```

```python
# Datos del ejercicio 1
xi1 = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi1 = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

# Resolver el ejercicio 1
ajustar_polinomios(xi1, yi1)
```

## **3. Promedios de puntos y calificaciones ACT**
La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemáticas y ciencias computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemáticas de la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos.

| Puntuación ACT | Promedio de puntos | Puntuación ACT | Promedio de puntos |
| -------------- | ------------------ | -------------- | ------------------ |
| 28             | 3.84               | 29             | 3.75               |
| 25             | 3.21               | 28             | 3.65               |
| 28             | 3.23               | 27             | 3.87               |
| 27             | 3.63               | 29             | 3.75               |
| 28             | 3.75               | 21             | 1.66               |
| 33             | 3.20               | 28             | 3.12               |
| 28             | 3.41               | 28             | 2.96               |
| 29             | 3.38               | 26             | 2.92               |
| 23             | 3.53               | 30             | 3.10               |
| 27             | 2.03               | 24             | 2.81               |

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
puntuaciones_ACT = np.array([28, 29, 25, 28, 27, 29, 28, 27, 28, 33, 28, 28, 29, 26, 23, 30, 24])
promedios_puntos = np.array([3.84, 3.75, 3.21, 3.65, 3.23, 3.87, 3.75, 3.41, 3.38, 3.20, 3.12, 2.96, 2.92, 3.53, 2.03, 3.10, 2.81])

# Calculamos la recta por mínimos cuadrados
A = np.vstack([puntuaciones_ACT, np.ones(len(puntuaciones_ACT))]).T
m, c = np.linalg.lstsq(A, promedios_puntos, rcond=None)[0]

# Graficamos los datos y la recta ajustada
plt.figure(figsize=(8, 6))
plt.scatter(puntuaciones_ACT, promedios_puntos, label='Datos')
plt.plot(puntuaciones_ACT, m*puntuaciones_ACT + c, 'r', label='Recta por mínimos cuadrados')
plt.xlabel('Puntuación ACT')
plt.ylabel('Promedio de puntos')
plt.title('Ajuste por mínimos cuadrados de datos de puntuaciones ACT y promedios de puntos')
plt.legend()
plt.grid(True)
plt.show()

# Mostramos los coeficientes de la recta ajustada
print(f"La recta ajustada es: y = {m:.2f}x + {c:.2f}")
```

## **4. Datos de supervivencia en choques de automóviles**
El siguiente conjunto de datos, presentado al Subcomité Antimonopolio del Senado, muestra las características comparativas de supervivencia durante un choque de automóviles de diferentes clases. Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria).

| Tipo                         | Peso promedio | Porcentaje de presentación |
| ---------------------------- | ------------- | -------------------------- |
| Regular lujoso doméstico     | 4800 lb       | 3.1                        |
| Regular intermediario doméstico | 3700 lb    | 4.0                        |
| Regular económico doméstico  | 3400 lb       | 5.2                        |
| Compacto doméstico           | 2800 lb       | 6.4                        |
| Compacto extranjero          | 1900 lb       | 9.6                        |

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
peso = np.array([4800, 3700, 3400, 2800, 1900])  # en libras
porcentaje = np.array([3.1, 4.0, 5.2, 6.4, 9.6])  # porcentaje

# Calculamos la recta por mínimos cuadrados
A = np.vstack([peso, np.ones(len(peso))]).T
m, c = np.linalg.lstsq(A, porcentaje, rcond=None)[0]

# Graficamos los datos y la recta ajustada
plt.figure(figsize=(8, 6))
plt.scatter(peso, porcentaje, label='Datos')
plt.plot(peso, m*peso + c, 'r', label='Recta por mínimos cuadrados')
plt.xlabel('Peso promedio (lb)')
plt.ylabel('Porcentaje de presentación (%)')
plt.title('Ajuste por mínimos cuadrados de datos de supervivencia en choques de automóviles')
plt.legend()
plt.grid(True)
plt.show()

# Mostramos los coeficientes de la recta ajustada
print(f"La recta ajustada es: y = {m:.2f}x + {c:.2f}")
```
